```python
import random
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
employee_ids=[f"E{str(i).zfill(3)}"for i in range(1,31)]
employee_ids
```

```python
working_hours_data={
    "EmployeeID":[],
    "Day":[],
    "HoursWorked":[]
}
working_hours_data
```

```python
for emp_id in employee_ids:
    for day in range(1,8):
        hours=random.uniform(4,10)
        working_hours_data["EmployeeID"].append(emp_id)
        working_hours_data["Day"].append(f"Day{day}")
        working_hours_data["HoursWorked"].append(round(hours,2))
```

```python
df=pd.DataFrame(working_hours_data)
df
```

```python
df.describe()
```

```python
grouped_by_employee=df.groupby("EmployeeID")["HoursWorked"].sum().reset_index(name="TotalHours")
grouped_by_employee
```

```python
grouped_by_employee.head()
```

```python
df["Anomaly"]=df["HoursWorked"].apply(lambda x:x<4.5 or x>9.5)
```

```python
anomalies=df[df["Anomaly"]==True]
```

```python
print("\n===Detected Anomalies===")
print(anomalies.head())
```
